```python
import matplotlib.pyplot as plt     #from matplotlib import pyplot as plt
%matplotlib inline

from matplotlib.ticker import MaxNLocator      #  for documentation
```

```python
year=[ 2014, 2015, 2016, 2017, 2018]
apl_price=[ 93.95, 112.15, 104.05 , 144.85, 169.49]
ms_price=[  39.01, 50.29, 57.05, 69.98, 94.39]

```

```python
plt.plot(year,apl_price,'--g',year,ms_price,'y')
plt.xlabel('year')
plt.ylabel('Stock price')
plt.axis([2013,2019, 20,180])
plt.show()
```

```python
fig_1=plt.figure(1,figsize=(20,5))      # declaring figure which consists of charts

apple_price=fig_1.add_subplot(121)      # chart_1  making   or    subplot1
microsoft_price=fig_1.add_subplot(122)   # chart_2  making   or   subplot2



apple_price.plot(year,apl_price)           # chart1 plot
microsoft_price.plot(year,ms_price,'--g')   # chart2 plot


apple_price.xaxis.set_major_locator(MaxNLocator(integer=True))   #chart_1 Formatting
microsoft_price.xaxis.set_major_locator(MaxNLocator(integer=True))  #chart_2 Formatting


plt.show()
```

```python
# How to make subplots/Charts quicker?

fig_2,axes=plt.subplots(1,2,figsize=(20,5))

axes[0].plot(year,apl_price)
axes[1].plot(year,ms_price,'--g')

plt.show()

```
